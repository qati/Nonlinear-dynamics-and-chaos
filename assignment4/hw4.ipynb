```python
import numpy as np
import random as rnd
```

## 1. feladat

```python
def iter1(x):
    r = rnd.random()
    if r<=1/3:
        n = np.zeros(2)
    elif r<=2/3:
        n = np.array([1,0])
    else:
        n = np.array([1/2, np.sqrt(3)/2])
    return 0.5*(x+n)
```

```python
x = np.array([0.2, 0.2])
xn = []
yn = []
for i in range(500000):
    x = iter1(x)
    xn.append(x[0])
    yn.append(x[1])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(0, (12,8))
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xn[1000:],yn[1000:], '.', markersize=0.4)
```

## 2. feladat

```python
def iter2(x):
    xnext = np.zeros(2, dtype=np.double)
    r = rnd.random()
    if r<=0.02:
        xnext[0] = 0.5
        xnext[1] = 0.27*x[1]
    elif r<=0.17:
        xnext[0] = -0.139*x[0] + 0.263*x[1] + 0.57
        xnext[1] =  0.246*x[0] + 0.224*x[1] - 0.036
    elif r<=0.3:
        xnext[0] =  0.17 *x[0] - 0.215*x[1] + 0.408
        xnext[1] =  0.222*x[0] + 0.176*x[1] + 0.0893
    else:
        xnext[0] =  0.781*x[0] + 0.034*x[1] + 0.1075
        xnext[1] = -0.032*x[0] + 0.739*x[1] + 0.27
    return xnext
```

```python
x = np.array([0.2, 0.2])
xn = []
yn = []
for i in range(1000000):
    x = iter2(x)
    xn.append(x[0])
    yn.append(x[1])
```

```python
plt.figure(0, (12,8))
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xn[1000:],yn[1000:], '.', markersize=0.3)
```
