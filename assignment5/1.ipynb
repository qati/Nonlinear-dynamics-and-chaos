```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def it(x,y):
    return np.mod(2*x+y, 1), np.mod(x+y, 1)
```

# Trajektoria

```python
x = [ np.random.uniform(0,1) ]
y = [ np.random.uniform(0,1) ]
for i in range(100000):
    xc = x[i]
    yc = y[i]
    xc, yc = it(xc, yc)
    x.append(xc)
    y.append(yc)
```

```python
plt.plot(x[100:],y[100:], '.', markersize=1)
```

# 1000 pontbol inditva, 10000 iterci utn a 1000 pont eloszlsa

```python
def itn(x, y, n):
    for i in range(n):
        x, y = it(x,y)
    return x,y
```

```python
x0 = np.random.uniform(0,1, 1000)
y0 = np.random.uniform(0,1, 1000)
```

## kezdeloszls

```python
plt.figure(0, (12,8))
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x0,y0, 'o')
```

## 10000 iterci utni eloszls

```python
x,y = itn(x0,y0, 10000)
```

```python
plt.figure(0, (12,8))

plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, 'o')
```

# Extra feladat: integrl

```python
def u(x,y):
    return x**2*3*y

def time_sum(n, x0, y0):
    x = x0
    y = y0
    s = 0.0
    for i in range(n):
        x,y = it(x,y)
        s += u(x,y)
    return s/float(n)
```

```python
x0 = np.random.uniform(0,1)
y0 = np.random.uniform(0,1)

ns = [10*i for i in range(1,1000)]
iv = []
for n in ns:
    iv.append(time_sum(n, x0, y0))
```

```python
plt.figure(0, (12,8))
plt.xlabel("n")
plt.ylabel(r"$\bar{u}$")
plt.plot(ns, iv, '.')
```
