```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def it(t, p, K):
    pn = np.mod(p+K*np.sin(t), 2*np.pi)
    return np.mod(t+pn, 2*np.pi), pn
```

```python
def plot(snp, n, K, skip=100):
    t = [np.random.uniform(0, 2*np.pi, snp)]
    p = [np.random.uniform(0, 2*np.pi, snp)]
    for i in range(n):
        tc = t[i]
        pc = p[i]
        tc, pc = it(tc, pc, K)
        t.append(tc)
        p.append(pc)
    tp = np.reshape(t, len(t)*len(t[0]))
    pp = np.reshape(p, len(p)*len(p[0]))
    plt.figure(0, (14,7))
    plt.xlabel("x", fontsize=18)
    plt.ylabel("y", fontsize=18)
    plt.plot(tp[skip:], pp[skip:], '.', markersize=0.05)
```

```python
plot(1000, 3000, 0.6)
```

```python
plot(1000, 3000, 0.9)
```

```python
plot(1000, 3000, 1.2)
```

```python
plot(1000, 3000, 2.0)
```
