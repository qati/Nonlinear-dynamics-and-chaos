```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def it(x,y,a,b):
    return (y+1-a*x**2),(b*x) 
```

```python
a=1.4
b=0.3
N=1000000
```

```python
x = [ np.random.uniform(0,1) ]
y = [ np.random.uniform(0,1) ]
for i in range(N):
    xc = x[i]
    yc = y[i]
    xc, yc = it(xc, yc, a, b)
    x.append(xc)
    y.append(yc)
```

```python
def plotxy(x,y, skip=100):
    plt.figure(0, (14,7))
    plt.xlabel("x", fontsize=18)
    plt.ylabel("y", fontsize=18)
    plt.plot(x[skip:],y[skip:], '.')
```

```python
plotxy(x,y)
```

# Extra feladat: zoomolas

```python
def itn_subrange(a,b, rngx, rngy, n):
    xc =  np.random.uniform(0,1)
    yc =  np.random.uniform(0,1)
    xr = []
    yr = []
    while True:
        xc, yc = it(xc, yc, a, b)
        if xc>=rngx[0] and xc<=rngx[1] and yc>=rngy[0] and yc<=rngy[1]:
            xr.append(xc)
            yr.append(yc)
        if len(xr)>n:
            break
    return xr, yr
```

```python
xr1, yr1 = itn_subrange(a,b, [-1.5, 1.5], [-0.4, 0.4], 10000)
```

```python
plotxy(xr1, yr1)
```

```python
xr2, yr2 = itn_subrange(a,b, [0.53, 0.73], [0.15,0.21], 10000)
```

```python
plotxy(xr2, yr2)
```

```python
xr3, yr3 = itn_subrange(a,b, [0.621, 0.641], [0.185,0.191], 10000)
```

```python
plotxy(xr3, yr3)
```
